```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mplt
import numpy as np
df = pd.read_excel('lianjia-11-15.xlsx')
%matplotlib inline
```

```python
df['列-1']
df.rename(columns = {
    '列-1':'title',
    '列-1-链接':'title-href',
    '列-2':'rent',
    '列-3':'community',
    '列-3-链接':'add-href',
    '列-4':'type',
    '列-5':'area',
    '列-6':'toward',
    '列-7':'address',
    '列-8':'subway',
    '列-9':'count',
},inplace = True)
df.drop(['列-7-链接','来源网址','add-href'],axis = 1,inplace=True)
# df
```

```python
def replace_value(value):
    value = value.replace('租房','')
    return value.replace('租房','')
df['address'] = df['address'].apply(replace_value)

def replace_value(value):
    value = value.replace('平米','')
    return value.replace('平米','')
df['area'] = df['area'].apply(replace_value)

df
```

```python
pd.value_counts(df['address']).head(25).plot.bar(
    figsize=(20,5),
    rot=45,
    fontsize=18,
    title:'房源数量'
)
```

```python
from operator import itemgetter
value = set( df['address'].values.tolist() )
rent_list = []

for item in value:
    flag = (df['address'] == item)
    rent = df[flag]['rent'].astype(int)
    area = df[flag]['area'].astype(int)
    rent_list.append({
        'value':round(rent.mean() / area.mean(),2), 
        'name':item
    })

rent_mean_list = sorted(rent_list,key = itemgetter('value'),reverse = True)
rent_mean_name = [item['name']  for item in rent_mean_list]
rent_mean_value = [item['value'] for item in rent_mean_list]

fig, ax = plt.subplots(figsize = (25,5))
bar_width = 0.5

x_index = np.arange(len(rent_mean_name))
x_index_width = 0.1

ax.bar(x_index,rent_mean_value,bar_width)
ax.set_title('每平米价格')
plt.yticks(fontsize = 18)
plt.xticks(x_index + x_index_width * 0.5, rent_mean_name,fontsize = 18,rotation = 40)
# ax.set_fontsize(18)
plt.show()
rent_mean_list
```
